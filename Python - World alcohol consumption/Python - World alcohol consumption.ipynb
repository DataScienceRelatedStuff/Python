```python
import pandas as pd
```

```python
a_c = pd.read_csv("World alcohol consumption dataset.csv")

a_c
```

```python
print("Shape:")
print(a_c.shape)
print("# Rows:")
print(a_c.shape[0])
print("# Columns:")
print(a_c.shape[1])
print("# Column Names:")
print(a_c.columns)
```

```python
print("First 10 rows:")
print("option 1:")
print(a_c.head(10))
print("option 2:")
print(a_c.iloc[:10])
```

```python
print("Display rows 54 till 60, columns 'Country' till 'Beverage Types':")
print(a_c.loc[55:59, "Country": "Beverage Types"])
print(a_c.loc[15:16,"Year":"Display Value"])
```

```python
print("Select first 4 rows with first 3 columns:")
print(a_c.iloc[:4,:3])
print("Display columns country & Beverage Type:")
print(a_c[['Country', 'Beverage Types']])
```

```python
print("Display 6 random records:")
print(a_c.sample(6))
print("Display 3% of random rows:")
print(a_c.sample(frac=0.03))
```

```python
print("Detect missing values:")
print(a_c.isna())
print("Remove/drop missing items:")
print(a_c.dropna())

print("Remove/drop duplicates:")
print(a_c.drop_duplicates())
```

```python
print("Select 5 first values from region Africa:")
print(a_c[a_c['WHO region'] == 'Africa'].head())

print("Select 5 first values from country Angola:")
print(a_c[a_c['Country'] == 'Angola'].head())


print("Select first 15 values from year 1989:")
print(a_c[
        a_c['Year'] == 1989
        ]
      .head(15))
```

```python
print("Select first 5 values from year 1987 or 1989:")
print((a_c
       [(a_c['Year'] == 1987) |
        (a_c['Year'] == 1989)
       ]
      ).head())
```

```python
print("Display first 5 values from region Africa AND country Democratic Republic of the Congo:")
print((a_c
       [
           (a_c['WHO region'] == 'Africa') &
           (a_c['Country'] == 'Democratic Republic of the Congo')
       ]
      ).head())

print("Display 5 first values from region Europe AND country Democratic Republic of the Congo:")
print((a_c
       [
           (a_c['WHO region'] == 'Europe') &
           (a_c['Country'] == 'Democratic Republic of the Congo')           
       ]
      ).head())
```

```python
print("Display first 5 values from region Africa AND country Niger from year 1984 or 1989:")
print((a_c
       [
           ((a_c['Year'] == 1984) |
           (a_c['Year'] == 1989)) &
           
           ((a_c['WHO region'] == 'Africa') &
           (a_c['Country'] == 'Niger'))  
       ]  
      ).head())

print("Display first 5 values from region Americas AND from year 1986 or 1988:")
print((a_c
       [
           ((a_c['Year'] == 1986) |
           (a_c['Year'] == 1988)) &
           
           (a_c['WHO region'] == 'Americas')   
       ]  
      ).head())

print("Display first 15 values from region Americas or Europe AND from year 1986 or 1989:")
print((a_c
       [
           ((a_c['Year'] == 1986) |
           (a_c['Year'] == 1989)) &
           
           ((a_c['WHO region'] == 'Americas') |
            (a_c['WHO region'] == 'Europe'))
       ]  
      ).head(15))
```

```python
print("Display 5 first values with 'Berverage Types' = 'Beer' AND 'Display Value' > 0.72:")
print(a_c
       [
           (a_c['Beverage Types'] == 'Beer') &
           (a_c['Display Value'] > 0.72)           
       ].head())

print("\nDisplay 5 first values with 'Berverage Types' = 'Wine' OR 'Beer' AND 'Display Value' < 0.72:")
print(a_c
       [
           ((a_c['Beverage Types'] == 'Wine') |
            (a_c['Beverage Types'] == 'Spirits'))
            &
           (a_c['Display Value'] < 0.72)           
       ].head())

print("\nDisplay 8 first values with 'Display Value' < 2.1 AND > 61:")
print(a_c[(a_c['Display Value'] < 2.1) & (a_c['Display Value']>.61)].head(8))

print("\nDisplay count of values with 'Display Value' < 2.1 AND > 61 in 'WHO region' 'Africa':")
print(a_c[(a_c['Display Value'] < 2.1) & (a_c['Display Value']>.61) & (a_c['WHO region'] == 'Africa')].count())
```

```python
print("Display the columns 'Country' and 'Beverage Types' for rows 2 till 7")
print(a_c.loc[1:6, ['Country', 'Beverage Types']])
```

```python
print("Remove all the na-items and store the result in new_a_c.")
new_a_c = a_c.dropna()
print("Display new_a_c.")
new_a_c

print('Display the items of countries that contain "ay" in their name:')
print(new_a_c[new_a_c['Country'].str.contains("ay")])

print('Display the items of countries "Norway", "Niger":')
new_a_c_alt = new_a_c[new_a_c['Country'].isin(["Norway", "Niger", "France"])]
print(new_a_c_alt)
```
