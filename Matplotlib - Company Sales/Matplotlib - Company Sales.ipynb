```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv("company_sales_data.csv")
sales_df.head()
```

```python
profit_list = sales_df['total_profit'].to_list()
month_list = sales_df['month_number'].to_list()
totalunits_list = sales_df['total_units'].to_list()

facecream_list = sales_df['facecream'].to_list()
facewash_list = sales_df['facewash'].to_list()
toothpaste_list = sales_df['toothpaste'].to_list()
bathingsoap_list = sales_df['bathingsoap'].to_list()
shampoo_list = sales_df['shampoo'].to_list()
moisturizer_list = sales_df['moisturizer'].to_list()

print('profit:', profit_list)
print('month:', month_list)
print('total units:', totalunits_list)
print('facecream:', facecream_list)
print('facewash:', facewash_list)
print('toothpaste:', toothpaste_list)
print('bathingsoap:', bathingsoap_list)
print('shampoo:', shampoo_list)
print('moisturizer:', moisturizer_list)
```
